<h2>What is Numpy?</h2>

* A Python library used for working with arrays.

* NumPy stands for Numerical Python.

* Most of the parts that require fast computation are written in C or C++.

* NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.

<h2>Array</h2>

* Continuous place in memory 

* A grid of values, all of the same type

* Indexed by a tuple of nonnegative integers

* rank: the number of dimensions (*Note: different with Linear ALgebra*)

* shape: the size of the array along each dimension

<h2>Scalar, vector, matrix, tensor </h2>

* Scalar $x\in \mathbb{R}$ : a 0th-order tensor

* Vector $x\in \mathbb{R}^{n \times 1}$ : a 1th-order tensor

* Matrix $x\in \mathbb{R}^{m \times n}$ : a 2th-order tensor

* Tensor : a nth-order tensor, e.g. (N, C, H, W)

這些物件在Numpy裡面都是稱作**array**，差別在維度!

<h2>Numpy array vs. Python list</h2>

* Example1: Multiply an array by 2

```python
# List comprehension
a = [1, 2, 3, 4, 5]
[2*q for q in a]
```

```python
# Numpy
a = np.array([1, 2, 3, 4, 5])
2*a
```

* Example2: Multiply the elements in two array (element-wise)

```python
a = [1, 2, 3, 4, 5]
b = [5, 4, 3, 2, 1]
# a*b
[q*r for q, r in zip(a,b)]
```

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 4, 3, 2, 1])
a * b
```

```python
import numpy as np
from PIL import Image

################## Read the image and numpy type conversion ########################
im  =  Image.open("test.jpg")  
img =  np.array(im)  # pixel value range [0, 255]
print(img.shape)     # (H, W, C) = (1706, 960, 3)    

################## Add Pepper Noise onto n pixel ######################################
n = int(1706*960/10)

for k in range(n):
        i = int(np.random.random() * img.shape[0]) # 0-axis (H)
        j = int(np.random.random() * img.shape[1]) # 1-axis (W)
        if img.ndim == 2:
            img[i,j] = 255
        elif img.ndim == 3:
            img[i,j,0]= 255
            img[i,j,1]= 255
            img[i,j,2]= 255

img = Image.fromarray(img.astype('uint8')).convert('RGB')

################## Show image ######################################
import matplotlib.pyplot as plt

imgplot = plt.imshow(img)
plt.show()
```
